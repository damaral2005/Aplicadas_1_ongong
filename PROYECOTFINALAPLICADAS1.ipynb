```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parmetros del modelo de Lotka-Volterra para conejos y zorros
a = 0.5   # Tasa de crecimiento de conejos
b = 0.02  # Tasa de encuentro entre conejos y zorros
c = 0.4   # Tasa de mortalidad de zorros
d = 0.01  # Eficiencia de conversin de conejos en descendencia de zorros

# Configuracin de tiempo
time = np.linspace(0, 200, 1000)  # Intervalo de tiempo para la simulacin

# Inicializacin de las poblaciones
initial_rabbits = 40  # Poblacin inicial de conejos
initial_foxes = 9     # Poblacin inicial de zorros

# Funcin del modelo de Lotka-Volterra
def lotka_volterra(state, t, a, b, c, d):
    x, y = state
    dxdt = a * x - b * x * y
    dydt = d * x * y - c * y
    return [dxdt, dydt]

# Resolucin de las ecuaciones diferenciales
solution = odeint(lotka_volterra, [initial_rabbits, initial_foxes], time, args=(a, b, c, d))
rabbits, foxes = solution.T  # Descomposicin de resultados en conejos y zorros

# Grfico de los resultados
plt.figure(figsize=(14, 6))

# Grfico de las poblaciones en el tiempo
plt.subplot(1, 2, 1)
plt.plot(time, rabbits, label="Conejos", color="blue")
plt.plot(time, foxes, label="Zorros", color="orange")
plt.title("Dinmica de poblaciones de conejos y zorros")
plt.xlabel("Tiempo")
plt.ylabel("Poblacin")
plt.legend()
plt.tight_layout()
plt.show()
```
