```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parámetros del modelo de Lotka-Volterra para conejos y zorros
a = 0.5   # Tasa de crecimiento de conejos
b = 0.02  # Tasa de encuentro entre conejos y zorros
c = 0.4   # Tasa de mortalidad de zorros
d = 0.01  # Eficiencia de conversión de conejos en descendencia de zorros

# Configuración de tiempo
time = np.linspace(0, 200, 1000)  # Intervalo de tiempo para la simulación

# Inicialización de las poblaciones
initial_rabbits = 40  # Población inicial de conejos
initial_foxes = 9     # Población inicial de zorros

# Función del modelo de Lotka-Volterra
def lotka_volterra(state, t, a, b, c, d):
    x, y = state
    dxdt = a * x - b * x * y
    dydt = d * x * y - c * y
    return [dxdt, dydt]

# Resolución de las ecuaciones diferenciales
solution = odeint(lotka_volterra, [initial_rabbits, initial_foxes], time, args=(a, b, c, d))
rabbits, foxes = solution.T  # Descomposición de resultados en conejos y zorros

# Gráfico de los resultados
plt.figure(figsize=(14, 6))

# Gráfico de las poblaciones en el tiempo
plt.subplot(1, 2, 1)
plt.plot(time, rabbits, label="Conejos", color="blue")
plt.plot(time, foxes, label="Zorros", color="orange")
plt.title("Dinámica de poblaciones de conejos y zorros")
plt.xlabel("Tiempo")
plt.ylabel("Población")
plt.legend()

# Diagrama de fase
plt.subplot(1, 2, 2)
plt.plot(rabbits, foxes, color="purple")
plt.title("Diagrama de fase: Conejos vs Zorros")
plt.xlabel("Conejos")
plt.ylabel("Zorros")

plt.tight_layout()
plt.show()
```
