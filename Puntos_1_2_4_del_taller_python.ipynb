```python
#Punto 1. fibonacci
import matplotlib.pyplot as plt
import numpy as np

def fibonacci(n):
    fib_series = [0, 1] #inicia en 0, 1
    for i in range(2, n):
        fib_series.append(fib_series[-1] + fib_series[-2]) # el -1 represente el numero anterior y el -2 representa el penultimo, como ya se sabe fibonacci es la suma de los 2 valores anteriores
    return fib_series

#@title Selecciona el valor de n { run: "auto", vertical-output: true }
n = 22  #@param {type:"slider", min:2, max:50, step:1}

fib_series = fibonacci(n)

plt.title("Serie de Fibonacci")
plt.xlabel("indice")
plt.ylabel("Valor Fibonacci")

plt.plot(range(n), fib_series, marker='o', color='green')

plt.grid()

plt.show()
```

```python
#punto 2 Meteorologico (decidi tomar datos de todos los dias del año porque me gusta complicarme la vida)
np.random.seed(1)  # funcion de numpy que te da numeros random
temperaturas = np.random.normal(loc=20, scale=5, size=360)  # loc es la media de las temperaturas y esta en 20 grados, scale representa el valor promedio de la desviacion estandar de 5 grados, size es la cantidad de valores generados, en este caso es 360, osea se generan 365 datos donde la mayoria son entre 15 a 25 grados celsius

media = np.mean(temperaturas)# np.mean es media
mediana = np.median(temperaturas)#np.median es la mediana
desviacion_estandar = np.std(temperaturas)#np.std te la la estandar

#el .2f representa la cantidad de decimales que queres mostrar, osea la aprox
print(f"media: {media:.2f} °C")
print(f"mediana: {mediana:.2f} °C")
print(f"desviacion estandar: {desviacion_estandar:.2f} °C")

# calcular la temperatura media mensual
temperaturas_mensuales = [np.mean(temperaturas[i:i+30]) for i in range(0, 360, 30)]


# Crear el gráfico de líneas de la temperatura media mensual
plt.figure(figsize=(10, 6))
plt.plot(range(1, 13), temperaturas_mensuales, marker='o')

plt.title("Temperatura Media Mensual")
plt.xlabel("Mes")
plt.ylabel("Temperatura Media (°C)")

# mostrar la cuadrícula
plt.grid()

# mostrar el gráfico
plt.show()
```

```python
#Punto 4 valores trigonometricos de 0 a 2 por pi
# mil valores entre 0 y 2 * pi
x = np.linspace(0, 2 * np.pi, 1000)

# calcular los valores de las funciones trigonometricas
y_seno = np.sin(x)
y_coseno = np.cos(x)
y_tangente = np.tan(x)

plt.figure(figsize=(10, 6))

plt.plot(x, y_seno, label='Seno', color='blue')
plt.plot(x, y_coseno, label='Coseno', color='green')
plt.plot(x, y_tangente, label='Tangente', color='red')

plt.title("Funciones Trigonométricas")
plt.xlabel("Ángulo (radianes)")
plt.ylabel("Valor")


plt.ylim(-2, 2)
plt.xlim(0, 2 * np.pi)

plt.grid()
plt.legend()

plt.show()
```
