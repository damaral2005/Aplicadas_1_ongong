Ejercicio 1:
$V = \pi r^{2} h$
$\frac{r}{R} = \frac{H-h}{H} \to r = R\left( 1 - \frac{h}{H} \right)$
Sustituimos:
$V = \pi \left(R\left(1 - \frac{h}{H}\right)\right)^{2} h = \pi R^{2} h \left( 1 - \frac{h}{H} \right)^{2}$
$V = \pi R^{2} h \left( 1 - \frac{2h}{H} + \frac{h^{2}}{H^{2}} \right) = \pi R^{2} \left( h - \frac{2h^{2}}{H} + \frac{h^{3}}{H^{2}} \right)$
Derivamos:
$\frac{dV}{dh} = \pi R^{2} \left( 1 - \frac{4h}{H} + \frac{3h^{2}}{H^{2}} \right)$
$1 - \frac{4h}{H} + \frac{3h^{2}}{H^{2}} = 0$
$H^{2} - 4Hh + 3h^{2} = 0$
$h = \frac{4H \pm \sqrt{(4H)^{2} - 4 \cdot 3 \cdot H^{2}}}{2 \cdot 3} = \frac{4H \pm \sqrt{16H^{2} - 12H^{2}}}{6} = \frac{4H \pm \sqrt{4H^{2}}}{6} = \frac{4H \pm 2H}{6}$
$h = H$
$h = \frac{H}{3}$
$r = R\left( 1 - \frac{1}{3} \right) = \frac{2R}{3}$
$r = \frac{2R}{3}, h = \frac{H}{3}$
$V = \pi \left( \frac{2R}{3} \right)^{2} \left( \frac{H}{3} \right) = \pi \frac{4R^{2}H}{27}$

Ejercicio 2:
$E = k\frac{d}{v + v_e}v^{3}$

$3v^{2}(v + v_e) - dv^{3} = 0$
$v^{2}(3(v + v_e) - dv) = 0$
$v = \frac{3v_e}{d - 3}$
$E_m = k\frac{d}{\frac{3v_e}{d - 3} + v_e \left( \frac{3v_e}{d - 3} \right)^{3}}$
