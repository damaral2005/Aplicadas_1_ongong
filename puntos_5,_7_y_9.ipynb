```python
#Punto a.
import matplotlib.pyplot as plt
import numpy as np
# Definimos la función a_n = (10^n) / n!
def a_n_factorial(n):
    return (10**n) / np.math.factorial(n)

# Calculamos los valores de a_n para n = 1, 2, ..., 100
n_range_factorial = np.arange(1, 101)
a_range_factorial = np.array([a_n_factorial(n) for n in n_range_factorial])

# Graficamos los puntos (n, a_n)
plt.figure(figsize=(10, 6))
plt.plot(n_range_factorial, a_range_factorial, label='$a_n = \\frac{10^n}{n!}$', color='green', marker='o')

#lindo
plt.title('Gráfico de $a_n = \\frac{10^n}{n!}$ para $n = 1, 2, \\ldots, 100$')
plt.xlabel('n')
plt.ylabel('$a_n$')
plt.legend()
plt.grid(True)
plt.show()
```

#Punto 9


Resolucion teorica:
>$a_1=2$,
>>$a_n+1=2a_n$


Esto significa que cada termino es el doble del anterior. Por lo tanto, podemos derivar una fórmula general para $an$ :
>$a_n = 2n$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función a_n
def a_n(n):
    return 2 ** n  # a_n = 2^n

# Generamos los valores de n
n_values = np.arange(1, 51)  # n de 1 a 50
a_n_values = [a_n(n) for n in n_values]

# Graficamos
plt.figure(figsize=(10, 6))
plt.plot(n_values, a_n_values, marker='o')
plt.title('Gráfica de la sucesión $a_n = 2^n$')
plt.xlabel('n')
plt.ylabel('$a_n$')
plt.yscale('log')  # Usamos escala logarítmica para visualizar mejor
plt.grid()
plt.show()
```

A medida que $n$ tiende a infinto $a_n$ tambien tiende a infinito: $lim_n->∞ a_n = ∞$
