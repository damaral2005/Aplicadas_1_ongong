```python
import numpy as np
import matplotlib.pyplot as plt


def rienman(n):
    fib_series = [0, 1] #inicia en 0, 1
    for i in range(1, n):
        fib_series.append(fib_series[-1] + fib_series[-2]) # el -1 represente el numero anterior y el -2 representa el penultimo, como ya se sabe fibonacci es la suma de los 2 valores anteriores
    return fib_series

n = 10


fib_series = rienman(n)

plt.title("Serie de rienman")
plt.xlabel("indice")
plt.ylabel("suma rienman")

plt.plot(range(n+1), fib_series, marker='o', color='green')

plt.grid()

plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_y(valores_naturales):
    return [(1 + 1/n)**n for n in valores_naturales]
n=50
valores_naturales = np.linspace(1,n,n)
y = calcular_y(valores_naturales)
e = np.e
plt.scatter(valores_naturales, y ,color = "blue" )
plt.axhline(y=e, linestyle="--", color="red")
plt.grid()
plt.show()
```
